# Data Cleaning

```python
data.MSRP.unique()
```

```python
data['MSRP'] = data['MSRP'].astype(str) #Changing the cost to string
data['MSRP'] = data['MSRP'].apply(lambda x: x.replace('$',''))
data['MSRP'] = data['MSRP'].apply(lambda x: x.replace(',','')) #Using lambda function to replace ',' from cost
data['MSRP'] = data['MSRP'].astype(float) # Changing the cost to Float
data.info() # looking at the dataset information after transformation
```
